```python
# 13792 종이붙이기 (DP 문제)
def ftn(n):
    k = n // 10
    if k == 1:
        return 1

    temp = 1
    for i in range(2, k + 1):
        if i % 2 == 1:
            temp = temp * 2 - 1
        else:
            temp = temp * 2 + 1

    return temp

T = int(input())

for test_case in range(1, T + 1):
    n = int(input())
    res = ftn(n)

    print(f"#{test_case} {res}")
```

```python
# 13789 괄호검사
def ftn(st):
    stack = []

    for i in st:
        if i == '(':
            stack.append(i)
        elif i == ')':
            if not stack:
                return 0
            elif stack.pop() != '(':
                return 0
        elif i == '{':
            stack.append(i)
        elif i == '}':
            if not stack:
                return 0
            elif stack.pop() != '{':
                return 0

    if stack:
        return 0

    return 1

T = int(input())

for test_case in range(1, T + 1):
    st = input()
    res = ftn(st)
```
