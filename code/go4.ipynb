```python
# 13569 gravity
T = int(input())

for test_case in range(1, T + 1):
    n = int(input())
    lst = list(map(int, input().split()))
    s = [0]*n

    for i in range(n):
        temp = lst[i]
        s_temp = 0
        for j in range(i + 1, n):
            if temp > lst[j]:
                s_temp += 1

        s[i] = s_temp

    mi = s[0]
    for i in range(n):
        if mi < s[i]:
            mi = s[i]

    print(f"#{test_case} {mi}")
```

```python
# 13568 구간합
def summa(x):
    s = 0

    for i in x:
        s += i

    return s


for test_case in range(1, T + 1):
    n, m = map(int, input().split())
    lst = list(map(int, input().split()))

    mi = summa(lst[0:m])
    for i in range(n - m + 1):
        temp = summa(lst[i:i + m])
        if mi > temp:
            mi = temp

    ma = summa(lst[0:m])
    for i in range(n - m + 1):
        temp = summa(lst[i:i + m])
        if ma < temp:
            ma = temp

    res = ma - mi
    print(f"#{test_case} {res}")
```

```python
# 13566 숫자 카드
T = int(input())
 
for test_case in range(1, T + 1):
    n = int(input())
    # lst = list(input())
    lst = input()
    dict1 = {}
 
    for i in lst:
        try:
            dict1[i] += 1
        except:
            dict1[i] = 1
 
    dict2 = list(sorted(dict1.items(), key = lambda x: (x[1], x[0]), reverse = True))
    res = dict2[0]
 
    print(f"#{test_case} {res[0]} {res[1]}")
```

```python
# 13565 전기버스
def ftn(k, n, lst):
    idx1 = 0
    cng = 0

    while idx1 + k < n:
        idx2 = idx1 + k
        road = range(idx1 + 1, idx2 + 1)
        st = []
        for i in lst:
            if i in road:
                st.append(i)

        if len(st) == 0:
            return 0

        cng += 1
        idx1 = st[-1]

    return cng


T = int(input())

for test_case in range(1, T + 1):
    k, n, m = map(int, input().split())
    lst = list(map(int, input().split()))

    res = ftn(k, n, lst)
    print(f"#{test_case} {res}")
```

```python
# 13564 min max
T = int(input())

for test_case in range(1, T + 1):
    n = int(input())
    lst = list(map(int, input().split()))

    mi = lst[0]
    ma = lst[0]

    for i in range(n):
        if mi > lst[i]:
            mi = lst[i]

        if ma < lst[i]:
            ma = lst[i]

    res = ma - mi

    print(f"#{test_case} {res}")
```
