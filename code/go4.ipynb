```python
# 13569 gravity
T = int(input())

for test_case in range(1, T + 1):
    n = int(input())
    lst = list(map(int, input().split()))
    s = [0]*n

    for i in range(n):
        temp = lst[i]
        s_temp = 0
        for j in range(i + 1, n):
            if temp > lst[j]:
                s_temp += 1

        s[i] = s_temp

    mi = s[0]
    for i in range(n):
        if mi < s[i]:
            mi = s[i]

    print(f"#{test_case} {mi}")
```

```python
# 13568 구간합
def summa(x):
    s = 0

    for i in x:
        s += i

    return s


for test_case in range(1, T + 1):
    n, m = map(int, input().split())
    lst = list(map(int, input().split()))

    mi = summa(lst[0:m])
    for i in range(n - m + 1):
        temp = summa(lst[i:i + m])
        if mi > temp:
            mi = temp

    ma = summa(lst[0:m])
    for i in range(n - m + 1):
        temp = summa(lst[i:i + m])
        if ma < temp:
            ma = temp

    res = ma - mi
    print(f"#{test_case} {res}")
```

```python
# 13565 전기버스
def ftn(k, n, lst):
    idx1 = 0
    cng = 0

    while idx1 + k < n:
        idx2 = idx1 + k
        road = range(idx1 + 1, idx2 + 1)
        st = []
        for i in lst:
            if i in road:
                st.append(i)

        if len(st) == 0:
            return 0

        cng += 1
        idx1 = st[-1]

    return cng


T = int(input())

for test_case in range(1, T + 1):
    k, n, m = map(int, input().split())
    lst = list(map(int, input().split()))

    res = ftn(k, n, lst)
    print(f"#{test_case} {res}")
```

```python
# import sys
# sys.stdin = open("input.txt", "r")

# 13704 달팽이 숫자
def pos(pre, num):
    num4 = num % 4
    direct = [[0, 1], [1, 0], [0, -1], [-1, 0]]  # 오른쪽, 아래, 왼쪽, 위
    a = pre[0] + direct[num4][0]
    b = pre[1] + direct[num4][1]
    res = [a, b]

    return res

T = int(input())

for test_case in range(1, T + 1):
    num = 0
    n = int(input())
    lst = [[0]*n for i in range(n)]
    start = [0, 0]

    for i in range(n**2):
        if (n == 1):
            lst[start[0]][start[1]] = i + 1
            break
        
        lst[start[0]][start[1]] = i + 1
        temp = pos(start, num)

        if temp[0] not in range(n) or temp[1] not in range(n):
            num = num + 1
            temp = pos(start, num)
        
        if lst[temp[0]][temp[1]] != 0:
            num = num + 1
            temp = pos(start, num)

        start = temp

    print(f"#{test_case}")
    for i in lst:
        print(*i)
```

```python
# import sys
# sys.stdin = open("input.txt", "r")

# 13702 델타 검색

def ab(n):
    if n < 0:
        return -n
    else:
        return n

def cross(lst, pt):
    lg = len(lst)
    direct = [
        [0, 0, -1, 1],
        [-1, 1, 0, 0]
        ]
    s = 0
    
    for i, j in zip(*direct):
        drow = pt[0] + i
        dcol = pt[1] + j

        if (drow >= 0 and drow < lg and dcol >= 0 and dcol < lg):
            temp1 = lst[pt[0]][pt[1]]
            temp2 = lst[drow][dcol]
            s += ab(temp2 - temp1)

    return s

T = 10

for test_case in range(1, T + 1):
    n = int(input())
    lst = [list(map(int, input().split())) for i in range(n)]
    total = 0

    for i in range(n):
        for j in range(n):
            total += cross(lst, [i, j])

    print(f"#{test_case} {total}")
```

```python
# 13700 이진탐색 (list1)
def bin(n, m):  # 전체 페이지 수, 찾으려는 페이지
    l = 1
    r = n
    c = int((l + r) / 2)
    cnt = 1

    while c != m:
        if c > m:
            r = c
            c = int((l + r) / 2)

        else:
            l = c
            c = int((l + r) / 2)

        cnt += 1

    return cnt


T = int(input())

for test_case in range(1, T + 1):
    p, a, b = map(int, input().split())
    na = bin(p, a)
    nb = bin(p, b)

    if na < nb:
        res = 'A'
    elif na == nb:
        res = 0
    else:
        res = 'B'

    print(f"#{test_case} {res}")
```

```python
# 13699 색칠하기

def clr(a, b):
    if a == 0 or b == 0:
        return a + b
    if a == b:
        return a

    return a + b


T = int(input())

for test_case in range(1, T + 1):
    n = int(input())
    total = 0
    field = [[0] * 10 for i in range(10)]

    for i in range(n):
        a, b, c, d, e = map(int, input().split())
        
        for j in range(a, c + 1):
            for k in range(b, d + 1):
                field[j][k] = clr(field[j][k], e)

    for j in range(10):
        for k in range(10):
            if field[j][k] == 3:
                total += 1

    print(f"#{test_case} {total}")
```

```python
# 1208 [S/W 문제해결 기본] 1일차 - Flatten

def find_max(lst):
    ma = lst[0]
    idx = 0

    for i in range(len(lst)):
        if (ma < lst[i]):
            ma = lst[i]
            idx = i

    return idx


def find_min(lst):
    ma = lst[0]
    idx = 0

    for i in range(len(lst)):
        if (ma > lst[i]):
            ma = lst[i]
            idx = i

    return idx


T = 10

for test_case in range(1, T + 1):
    n = int(input())
    lst = list(map(int, input().split()))

    for i in range(n):
        max_idx = find_max(lst)
        min_idx = find_min(lst)

        if (lst[max_idx] - lst[min_idx] <= 1):
            break

        lst[max_idx] -= 1
        lst[min_idx] += 1

    max_idx = find_max(lst)
    min_idx = find_min(lst)
    res = lst[max_idx] - lst[min_idx]

    print(f"#{test_case} {res}")
```

```python
# 1945 간단한 소인수분해
n = [2, 3, 5, 7, 11]

T = int(input())

for test_case in range(1, T + 1):
    num = int(input())
    lst = [0] * 5

    for i in range(5):

        while num % n[i] == 0:
            lst[i] += 1
            num //= n[i]

    print(f"#{test_case}", end = ' ')
    print(*lst)
```
