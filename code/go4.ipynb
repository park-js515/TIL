```python
# 1216 [S/W 문제해결 기본] 3일차 - 회문2
import copy

# 전치
def transpose(lst):
    temp = copy.deepcopy(lst)

    for i in range(100):
        for j in range(100):
            if i > j:
                temp[i][j], temp[j][i] = temp[j][i], temp[i][j]

    return temp

# 한 개를 기준의 대칭인지 확인한다.
# 두 개를 기준으로 대칭인지 확인한다. (두 개  같다면)

def check1(lst):
    res = 0

    for i in range(100):
        left = i - 1
        right = i + 1
        cnt = 1

        while 0 <= left < 100 and 0 <= right < 100:
            if lst[left] == lst[right]:
                cnt += 2
                left -= 1
                right += 1
            else:
                break

        if res < cnt:
            res = cnt

    return res

def check2(lst):
    res = 0

    for i in range(1, 100):
        if lst[i - 1] == lst[i]:
            left = i - 2
            right = i + 1
            cnt = 2
        else:
            continue

        while 0 <= left < 100 and 0 <= right < 100:
            if lst[left] == lst[right]:
                cnt += 2
                left -= 1
                right += 1
            else:
                break

        if res < cnt:
            res = cnt

    return res


T = 10
for test_case in range(1, T + 1):
    n = int(input())
    lst = [list(input()) for i in range(100)]
    lst2 = transpose(lst)
    res = 0

    for i in range(100):
        temp1 = check1(lst[i])
        temp2 = check1(lst2[i])
        temp3 = check2(lst[i])
        temp4 = check2(lst2[i])

        temp = max([temp1, temp2, temp3, temp4])

        if res < temp:
            res = temp


    print(f"#{test_case} {res}")
```

```python
# 1209. [S/W 문제해결 기본] 2일차 - Sum
def mx(lst):
    ma = 0

    for i in lst:
        if ma < sum(i):
            ma = sum(i)

    for i in range(100):
        k = 0
        for j in range(100):
            k += lst[j][i]

        if ma < k:
            ma = k

    k = 0
    for i in range(100):
        k += lst[i][i]

    if ma < k:
        ma = k

    k = 0
    for i in range(100):
        k += lst[i][99 - i]

    if k > ma:
        ma = k

    return ma


T = 10
for test_case in range(1, T + 1):
    _ = int(input())
    lst = [list(map(int, input().split())) for i in range(100)]
    temp = mx(lst)

    print(f"#{test_case} {temp}")
```
