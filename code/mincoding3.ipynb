```python
# 김밥 속재료

def bfs(level, start, s):
    global st

    if level == 3:
        print(s)
        return

    for i in range(start, len(st)):
        bfs(level + 1, i, s + st[i])

st = input()
bfs(0, 0, "")
```

```python
# Best Number 선정하기

def calc(l):
    return l[0] * l[1] - l[2] * l[3] + l[4]

def dfs(level, lst_now):
    global lst, visited, mi, ma

    if level == 5:
        if ma < calc(lst_now):
            ma = calc(lst_now)
        if mi > calc(lst_now):
            mi = calc(lst_now)
        return

    for i in range(5):
        if not visited[i]:
            visited[i] = 1
            dfs(level + 1, lst_now + [lst[i]])
            visited[i] = 0

lst = list(map(int, input().split()))
visited = [0] * 5
mi = 21e8
ma  = - 21e8
dfs(0, [])

print(ma)
print(mi)
```

```python
# 가수 BTS와 SES

def dfs(a, temp_cnt):
    global lst, st, cnt

    if len(a) == len(st):
        if cnt > temp_cnt:
            cnt = temp_cnt
        return

    for i in lst:
        temp = a + i

        if temp == st[:len(temp)]:
            dfs(temp, temp_cnt + 1)
            break

lst = ["BTS", "SBS", "BS", "CBS", "SES"]
st = input()
cnt = 21e8
dfs("", 0)
print(cnt)
```

```python
# 최소 동전 교환기

def dfs(s, cnt):
    global coins, changes, res

    if cnt > res:
        return

    if s == changes:
        if res > cnt:
            res = cnt
        return

    for i in coins:
        if s + i <= changes:
            dfs(s + i, cnt + 1)

coins = [10, 40, 60]
changes = int(input())
res = 21e8
dfs(0, 0)
print(res)
```

```python
# 10을 만들자

def dfs(level, s):
    global k, cnt

    if level == k:
        if s == 10:
            cnt += 1
        return

    for i in range(1, 10):
        if s + i <= 10:
            dfs(level + 1, s + i)

k = int(input())
cnt = 0
dfs(0, 0)
print(cnt)
```

```python
# 다빈치 민코드

def mul(nums):
    ret = 1

    for i in nums:
        ret *= i

    return ret

def dfs(level, s):
    global n, m, lst, visited, mi, res

    if level == m:
        if mi > mul(s):
            mi = mul(s)
            res = s
        return

    for i in range(n):
        if not visited[i]:
            visited[i] = 1
            dfs(level + 1, s + [lst[i]])
            visited[i] = 0


n, m = map(int, input().split())
lst = list(map(int, input().split()))
visited = [0] * n
mi = 21e8
res = []
dfs(0, [])
res.sort()

print(*res)
```

```python
# 구매팀 결재

def calc(lst):
    global chg

    s = 0

    for i in lst:
        s += chg.get(i)

    return s

def dfs(level, s):
    global abc, price, chg, st, n, cnt, ma

    if level == n:
        if s % 10 == 0:
            if ma < s:
                ma = s
        return

    for i in range(7):
        if cnt[i]:
            cnt[i] -= 1
            dfs(level + 1, s - chg.get(abc[i]))
            cnt[i] += 1

abc = "abcdefg"
price = [15, 20, 44, 22, 55, 16, 45]
chg = {key: value for key, value in zip(abc, price)}
chg2 = {key: value for value, key in enumerate(abc)}

st = input()
n = int(input())
ma = 0
cnt = [0] * len(abc)
for i in st:
    cnt[chg2.get(i)] += 1

dfs(0, calc(st))
print(ma)
```

```python
# 레드 마운틴

def dfs(v1, v2):
    global d, n, field, visited, res

    visited[v1][v2] = 1

    if v1 == n - 1 and v2 == n - 1:
        res = True
        return

    for i in range(4):
        drow = v1 + d[i][0]
        dcol = v2 + d[i][1]

        if 0 <= drow < n and 0 <= dcol < n:
            if not field[drow][dcol] and not visited[drow][dcol]:
                dfs(drow, dcol)
                visited[drow][dcol] = 0


d = ((-1, 0), (1, 0), (0, -1), (0, 1))
n = int(input())
field = [list(map(int, input().split())) for _ in range(n)]
visited = [[0] * n for _ in range(n)]
res = False
dfs(0, 0)

if res:
    print("가능")
else:
    print("불가능")
```

```python
# 가로 큐브 돌리기

import copy

def score(lst):
    lst_t = list(map(list, zip(*lst)))
    lst_t = list(map(sum, lst_t))
    s = 1

    for i in lst_t:
        s *= i

    return s

def cir(lst, k):
    temp = lst[:]

    for i in range(k):
        a = temp.pop()
        temp = [a] + temp

    return temp

def dfs(level, now):
    global n, ma

    if level == n:
        if ma < score(now):
            ma = score(now)
        return

    temp_now = copy.deepcopy(now)

    for i in range(n):
        temp_now[level] = cir(temp_now[level], i)
        dfs(level + 1, temp_now)

n = int(input())
cube = [list(map(int, input().split())) for _ in range(n)]
ma = -21e8

dfs(0, cube)
print(f"{ma}점")
```

```python
# 폭탄 투하 장소 선정

import copy

def boom(lst, r, c):

    for i in range(r - 1, r + 2):
        for j in range(c - 1, c + 2):
            if 0 <= i < 4 and 0 <= j < 4:
                if abs(r - i) + abs(c - j) <= 1:
                    lst[i][j] = '_'

def cnt_alp(lst):
    cnt = 0

    for i in range(4):
        for j in range(4):
            if lst[i][j] != '_':
                cnt += 1

    return cnt

def dfs(level, now, alp):
    global n, mi, res

    if level == n:
        if mi > cnt_alp(now):
            mi = cnt_alp(now)
            res = alp
        return

    temp_now = copy.deepcopy(now)
    for i in range(4):
        for j in range(4):
            temp_now2 = copy.deepcopy(temp_now)

            if temp_now2[i][j] != '_':
                temp_alp = temp_now2[i][j]
                boom(temp_now2, i, j)
                dfs(level + 1, temp_now2, alp + [temp_alp])


field = [list(input()) for _ in range(4)]
n = int(input())
mi = 21e8
res = []
dfs(0, field, [])
res.sort()
print(*res)
```

```python
# 수학이의 연산자

def calc(a, b, tp):
    if tp == 0:
        return (a - b) * (a + b)
    elif tp == 1:
        return max(a, b)
    elif tp == 2:
        return (a ** 2) - (b ** 2)
    elif tp == 3:
        return (a + b) ** 2

def stack_calc(step):
    global lst

    stack = lst[::-1]

    for i in step:
        a = stack.pop()
        b = stack.pop()

        stack.append(calc(a, b, i))

    return stack[0]


def dfs(level, step):
    global n, lst, cnt

    if level == n - 1:
        if stack_calc(step) == 100:
            cnt += 1
        return

    for i in range(4):
        dfs(level + 1, step + [i])


n = int(input())
lst = list(map(int, input().split()))
cnt = 0

dfs(0, [])
print(cnt)
```

```python
# 불화살 쏘기

def shot(target, sh):
    global n

    for i in range(sh - 1, sh + 2):
        if 0 <= i < n:
            target[i] = 0

def dfs(target, score):
    global n, ma, s

    if sum(target) == 0:
        if ma < sum(score):
            ma = sum(score)
            s = score
        return

    now_target = target[:]
    
    for i in range(n):
        now_target2 = now_target[:]

        if now_target2[i]:
            temp = now_target2[i]
            shot(now_target2, i)
            dfs(now_target2, score + [temp])


n = int(input())
lst = list(map(int, input().split()))
ma = -21e8
s = []

dfs(lst, [])
s2 = list(map(str, s))
print("+".join(s2), "=", sum(s), sep = '')
```

```python
# 독수리 3형제

def mul(lst):
    temp = list(map(lambda x: 2 * x, lst))

    return temp

def dfs(level, feed, s):
    global n, ma

    if level == n:
        if ma < s:
            ma = s
        return

    now_feed = feed[:]

    for i in range(3):
        for j in range(3, 6):
            for k in range(1, 5):
                now_feed2 = now_feed[:]
                temp = 0

                temp += now_feed2[i]
                now_feed2[i] = 0

                temp += now_feed2[j]
                now_feed2[j] = 0

                temp += now_feed2[k]
                now_feed2[k] = 0

                now_feed2 = mul(now_feed2)
                dfs(level + 1, now_feed2, s + temp)


lst = list(map(int, input().split()))
n = int(input())
ma = -21e8

dfs(0, lst, 0)
print(ma)
```

```python
# 디자이너의 손길

def score(lst):
    global lg

    sc = 0

    for i in range(lg - 1):
        if lst[i] == lst[i + 1]:
            sc -= 50
        elif abs(ord(lst[i]) - ord(lst[i + 1])) <= 5:
            sc += 3
        elif abs(ord(lst[i]) - ord(lst[i + 1])) >= 20:
            sc += 10

    return sc


def swap(lst, i, j):
    lst[i], lst[j] = lst[j], lst[i]


def dfs(level, cali):
    global n, lg, best

    if level == n:
        temp = score(cali)
        if best < temp:
            best = temp
        return

    temp_cali = cali[:]

    for i in range(lg):
        for j in range(i + 1, lg):
            temp_cali2 = temp_cali[:]
            swap(temp_cali2, i, j)

            dfs(level + 1, temp_cali2)


st = list(input())
lg = len(st)
n = int(input())
best = -21e8

dfs(0, st)
print(best)
```

```python
# 디자이너의 손길(2)

def score(lst):
    global lg

    sc = 0

    for i in range(lg - 1):
        if lst[i] == lst[i + 1]:
            sc -= 50
        elif abs(ord(lst[i]) - ord(lst[i + 1])) <= 5:
            sc += 3
        elif abs(ord(lst[i]) - ord(lst[i + 1])) >= 20:
            sc += 10

    return sc


def swap(lst, i, j):
    lst[i], lst[j] = lst[j], lst[i]


def dfs(level, cali):
    global n, lg, best

    if level == n:
        temp = score(cali)
        if best < temp:
            best = temp
        return

    temp_cali = cali[:]

    for i in range(lg):
        for j in range(i + 1, lg):

            swap(temp_cali, i, j)
            dfs(level + 1, temp_cali)
            swap(temp_cali, i, j) # 되돌리기


st = list(input())
lg = len(st)
n = int(input())
best = -21e8

dfs(0, st)
print(best)
```
