```python
# 14503 로봇 청소기

from collections import deque

def bfs(row, col, now_d):
    global d, chg_rev, chg_ort, r, c, field, cnt

    q = deque()
    q.append((row, col, now_d))

    while q:
        row, col, now_d = q.popleft()

        if not field[row][col]:
            cnt += 1
        field[row][col] = 2
        temp_cnt = 0

        for i in range(4):
            drow = row + d[i][0]
            dcol = col + d[i][1]

            if not field[drow][dcol]:
                temp_cnt += 1
                break

        if not temp_cnt:
            drow = row + d[chg_rev.get(now_d)][0]
            dcol = col + d[chg_rev.get(now_d)][1]

            if field[drow][dcol] != 1:
                q.append((drow, dcol, now_d))

        else:
            while 1:
                now_d = chg_ort.get(now_d)
                drow = row + d[now_d][0]
                dcol = col + d[now_d][1]

                if not field[drow][dcol]:
                    q.append((drow, dcol, now_d))
                    break


d = ((-1, 0), (0, 1), (1, 0), (0, -1))
chg_rev = {0: 2, 1: 3, 2: 0, 3: 1} # 반대
chg_ort = {0: 3, 1: 0, 2: 1, 3: 2} # 반시계 90도
r, c = map(int, input().split())
row, col, now_d = map(int, input().split())
field = [list(map(int, input().split())) for _ in range(r)]
cnt = 0

bfs(row, col, now_d)
print(cnt)
```
