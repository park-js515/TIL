```python
# 55 1948 임계경로

from collections import deque
def bfs(s):
    global adj_lst, dag, cp, n

    queue = deque()
    queue.append(s)

    while queue:
        now = queue.popleft()

        for i in adj_lst[now]:
            dag[i[0]] -= 1
            cp[i[0]] = max(cp[i[0]], cp[now] + i[1])

            if dag[i[0]] == 0:
                queue.append(i[0])

def bfs2(e):
    global adj_lst_rev, cp, n

    visited = [0] * (n + 1)
    queue = deque()
    queue.append(e)
    cnt = 0

    while queue:
        now = queue.popleft()

        for i in adj_lst_rev[now]:

            if cp[now] == cp[i[0]] + i[1]:
                cnt += 1

                if not visited[i[0]]:
                    visited[i[0]] = 1
                    queue.append(i[0])

    return cnt

n = int(input())
m = int(input())
adj_lst = [[] for _ in range(n + 1)]
adj_lst_rev = [[] for _ in range(n + 1)]
dag = [0] * (n + 1)
cp = [0] * (n + 1)

for _ in range(m):
    s, e, d = map(int, input().split())
    adj_lst[s].append([e, d])
    adj_lst_rev[e].append([s, d])
    dag[e] += 1

s, e = map(int, input().split())
bfs(s)
cnt = bfs2(e)

print(cp[e])
print(cnt)
```

```python
# 55 1948 임계경로
# 큐에 집어 넣을 때는 되도록 튜플로 넣자. 속도 차이가 많이 있다.(20%)

from collections import deque
def bfs(s, e):
    global adj_lst, dag, cp, n

    queue = deque()
    queue.append(s)

    while queue:
        now = queue.popleft()

        for i in adj_lst[now]:
            dag[i[0]] -= 1
            cp[i[0]] = max(cp[i[0]], cp[now] + i[1])

            if dag[i[0]] == 0:
                queue.append(i[0])

    visited = [0] * (n + 1)
    queue.clear()
    queue.append(e)
    cnt = 0

    while queue:
        now = queue.popleft()

        for i in adj_lst_rev[now]:

            if cp[now] == cp[i[0]] + i[1]:
                cnt += 1

                if not visited[i[0]]:
                    visited[i[0]] = 1
                    queue.append(i[0])

    return cnt


n = int(input())
m = int(input())
adj_lst = [[] for _ in range(n + 1)]
adj_lst_rev = [[] for _ in range(n + 1)]
dag = [0] * (n + 1)
cp = [0] * (n + 1)

for _ in range(m):
    s, e, d = map(int, input().split())
    adj_lst[s].append((e, d))
    adj_lst_rev[e].append((s, d))
    dag[e] += 1

s, e = map(int, input().split())

cnt = bfs(s, e)

print(cp[e])
print(cnt)
```

```python
# 56 1753 최단 경로 구하기(최단경로)

import heapq
def dijkstra(start):
    global adj_lst, v

    visited = [0] * (v + 1)
    distance = [1e8] * (v + 1)

    pq = []
    heapq.heappush(pq, (0, start))
    distance[start] = 0

    while pq:
        _, now = heapq.heappop(pq)

        if visited[now]:
            continue

        visited[now] = 1

        for nt, dist in adj_lst[now]:
            if distance[nt] > distance[now] + dist:
                distance[nt] = distance[now] + dist
                heapq.heappush(pq, (distance[nt], nt)) # 변화가 있을 때만 집어 넣는다.

    return visited, distance

v, e = map(int, input().split())
k = int(input())
adj_lst = [[] for _ in range(v + 1)]

for i in range(e):
    a, b, d = map(int, input().split())
    adj_lst[a].append((b, d))

visited, distance = dijkstra(k)

for i in range(1, v + 1):
    if visited[i]:
        print(distance[i])
    else:
        print("INF")
```

```python
# 57 1916 최소비용 구하기
# PriorityQueue 버전
import sys
from queue import PriorityQueue
def dijkstra(start):
    global adj_lst, distance, visited, n

    pq = PriorityQueue()
    pq.put((0, start))
    distance[start] = 0

    while pq.qsize() > 0:
        temp = pq.get()
        now = temp[1]

        if visited[now]:
            continue
        visited[now] = 1

        for i in adj_lst[now]:
            if distance[i[0]] > distance[now] + i[1]:
                distance[i[0]] = distance[now] + i[1]
                pq.put((distance[i[0]], i[0]))

n = int(input())
m = int(input())
adj_lst = [[] for _ in range(n + 1)]
distance = [sys.maxsize] * (n + 1)
visited = [0] * (n + 1)

for _ in range(m):
    s, e, d = map(int, input().split())
    adj_lst[s].append((e, d))

s, e = map(int, input().split())
dijkstra(s)

print(distance[e])
```

```python
# 57 1916 최소비용 구하기
# heapq 버전 (메모리, 시간 측면에서 모두 우위에 있다.)

import sys
import heapq
def dijkstra(start):
    global adj_lst, distance, visited

    pq = []
    heapq.heappush(pq, (0, start))
    distance[start] = 0

    while pq:
        temp = heapq.heappop(pq)
        now = temp[1]

        if visited[now]:
            continue
        visited[now] = 1

        for i in adj_lst[now]:
            if distance[i[0]] > distance[now] + i[1]:
                distance[i[0]] = distance[now] + i[1]
                heapq.heappush(pq, (distance[i[0]], i[0]))

n = int(input())
m = int(input())
adj_lst = [[] for _ in range(n + 1)]
distance = [sys.maxsize] * (n + 1)
visited = [0] * (n + 1)

for _ in range(m):
    s, e, d = map(int, input().split())
    adj_lst[s].append((e, d))

s, e = map(int, input().split())
dijkstra(s)

print(distance[e])
```

```python
# 58 1854 k 번째 최단 경로 찾기

import heapq
def dijkstra():
    global adj_lst, distance

    pq = []
    heapq.heappush(pq, (0, 1))
    distance[1][0] = 0

    while pq:
        now_cost, now = heapq.heappop(pq)

        for nxt, nxt_cost in adj_lst[now]:
            if distance[nxt][k - 1] > now_cost + nxt_cost:
                distance[nxt][k - 1] = now_cost + nxt_cost
                distance[nxt].sort()
                heapq.heappush(pq, (now_cost + nxt_cost, nxt))

n, m, k = map(int, input().split())
adj_lst = [[] for _ in range(n + 1)]
distance = [[sys.maxsize] * k for _ in range(n + 1)]

for _ in range(m):
    a, b, c = map(int, input().split())
    adj_lst[a].append((b, c))

dijkstra()

for i in range(1, n + 1):
    if distance[i][k - 1] == sys.maxsize:
        print(-1)
    else:
        print(distance[i][k - 1])
```
