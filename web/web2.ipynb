- 모든 요소는 네모(박스모델)이고, 좌측상단에 배치.  
- <font color="red">display에 따라 크기와 배치가 달라진다.<font>

<b>대표적으로 활용되는 display</b>  

display: block  
- 줄바꿈이 일어나는 요소(다른 element를 밀어낸다!)  
- 화면 크기 전체의 가로 폭을 차지한다.  
- 블록 레벨 요소 안에 인라인 레벨 요소가 들어갈 수 있음.  
- 대표적인 블록 레벨 요소: div/ ul, ol, li/ p / hr/ form 등
- <a href="https://developer.mozilla.org/ko/docs/Web/HTML/Block-level_elements"> link </a>


&nbsp;

display: inline  
- 줄바꿈이 일어나지 않는 행의 일부 요소  
- content를 마크업하고 있는 만큼 가로 폭을 차지한다.  
- width, height, margin-top, margin-bottom을 지정할 수 없다.  
- 상하의 여백은 line-height로 지정한다.(중요!)  
- 대표적인 인라인 레벨 요소: span / img / input, label / b, em, i, strong 등
- <a href="https://developer.mozilla.org/ko/docs/Web/HTML/Inline_elements"> link </a>

<b>display</b>  

display: inline-block  
- block과 inline 레벨 요소의 특징을 모두 가짐  
- inlne처럼 한줄에 표시 가능하고, block처럼 width, height, marign 속성을 모두 지정할 수 있음  

&nbsp;

display: none  
- 해당 요소를 화면에 표시하지 않고, 공간조차 부여되지 않음
- 이와 비슷한 visibility: hidden은 해당 요소가 공간은 차지하나 화면에 표시만 하지 않는다.  
- <a href="https://developer.mozilla.org/ko/docs/Web/CSS/display">link</a>

### CSS position

- 문서 상에서 요소의 위치를 지정 (어떤 기준으로 어디에 배치시킬지)
- <a href="https://developer.mozilla.org/ko/docs/Web/CSS/position">link</a>
- static: 모든 태그의 기본값 (기준 위치)
    - 일반적인 요소의 배치 순서 따름(normal flow: 좌측 상단)
    - 부모 요소 내의 배치될 때는 부모 요소의 위치를 기준으로 배치 됨  

&nbsp;

- 아래는 좌표 property(top, bottom, left, right)를 사용하여 이동 가능  

&nbsp;

- relative: 상대 위치
    - 자기 자신의 static 위치를 기준으로 이동(normal flow 유지)
    - 레이아웃에서 요소가 차지하는 공간은 static 일 때와 같음 (noraml position 대비 offset)

&nbsp;

- absolute: 절대 위치
    - 요소를 일반적인 문서 흐름에서 제거 후 레이아웃에서 공간을 차지하지 않음 (normal flow에서 벗어남)
    - static이 아닌 가장 가까이 있는 부모/조상 요소를 기준으로 이동 (없는 경우 body)

&nbsp;

- fixed: 고정 위치
    - 요소를 일반적인 문서 흐름에서 제거 후 레이아웃에서 공간을 차지하지 않음(normal flow에서 벗어남)
    - 부모 요소와 관계없이 viewport(실제로 보는 화면)를 기준으로 이동
        - 스크롤 시에도 항상 같은 곳에 위치함

&nbsp;

- sticky: 스크롤에 따라 static -> fixed로 변경  
    - 속성을 적용한 박스는 평소에 문서 안에서 position: static 상태와 같이 일반적인 흐름에 따르지만, 스크롤 위치가 **임계점**에 이르면 position: fixed와 같이 박스를 화면에 고정할 수 있는 속성
